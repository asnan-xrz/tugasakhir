# Visualisasi Praproses Data Citra
Sesuai dengan Kode Sumber 3.2 untuk implementasi praproses data citra, notebook ini menampilkan visualisasi gambar asli dan gambar setelah melalui serangkaian tahap praproses dan normalisasi untuk melengkapi Tabel 3.1.

```python
import torch
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
import os

# Parameter resolusi (sesuaikan jika berbeda, misal 224, 256, atau 512)
resolution = 256
interpolation = transforms.InterpolationMode.BILINEAR

# Implementasi Kode Sumber 3.2 (pengacakan/random dimatikan untuk visualisasi statis Tabel 3.1)
train_transforms = transforms.Compose([
    transforms.Resize(resolution, interpolation=interpolation),
    transforms.CenterCrop(resolution),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5]),
])

image_filenames = [
    "Q004.jpg",
    "Q019.jpg",
    "Q027.jpg",
    "Q030.jpg",
    "Q202.jpg"
]

# Direktori dataset gambar
base_dir = "allaboutITS"

# Siapkan plot untuk Tabel 3.1
fig, axes = plt.subplots(len(image_filenames), 2, figsize=(10, 20))
fig.suptitle('Tabel 3.1 Contoh hasil visual sebelum dan sesudah normalisasi', fontsize=16, y=1.02)

for i, filename in enumerate(image_filenames):
    img_path = os.path.join(base_dir, filename)
    img = Image.open(img_path).convert('RGB')
    
    # --- Kolom 1: Gambar Asli ---
    ax_orig = axes[i, 0]
    ax_orig.imshow(img)
    ax_orig.set_title(f"{i+1}. Gambar Asli ({filename})")
    ax_orig.axis('off')
    
    # --- Kolom 2: Gambar Hasil Normalisasi ---
    # Aplikasikan transformasi tensor
    tensor_img = train_transforms(img)
    
    # Denormalisasi balik dari [-1, 1] ke [0, 1] agar bisa divisualisasikan oleh matplotlib
    # Ini perlu agar gambar tidak berwarna aneh atau terpotong/clipping saat ditampilkan
    tensor_img_vis = (tensor_img * 0.5) + 0.5
    tensor_img_vis = tensor_img_vis.clamp(0, 1)
    
    # Pindah dimensi channel dari (C, H, W) menjadi (H, W, C) untuk plotting
    plot_img = tensor_img_vis.permute(1, 2, 0).numpy()
    
    ax_norm = axes[i, 1]
    ax_norm.imshow(plot_img)
    ax_norm.set_title(f"Gambar Hasil Normalisasi ({filename})")
    ax_norm.axis('off')

plt.tight_layout()
plt.show()

# Opsional: Simpan hasil plot ke dalam satu file gambar
fig.savefig('tabel_3_1_visualisasi_normalisasi.png', bbox_inches='tight', dpi=300)
print("Visualisasi berhasil disimpan sebagai tabel_3_1_visualisasi_normalisasi.png")
```
